## Agent Memory

```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ["GOOGLE_API_KEY"]=os.getenv("GOOGLE_API_KEY")
```

```python
from langchain.agents import create_agent
from langgraph.checkpoint.memory import InMemorySaver
```

```python
agent = create_agent(
    model="google_genai:gemini-2.5-flash-lite",
    checkpointer=InMemorySaver() # An in-memory checkpoint saver. This checkpoint saver stores checkpoints in memory using a defaultdict.
)
```

```python
# Run with thread id
config = {"configurable": {"thread_id": "test1"}}
```

```python
from langchain_core.messages import HumanMessage, SystemMessage
response = agent.invoke({"messages":[HumanMessage(content="Hi my name is Chinmay")]},config=config)
response
```

```python
agent.invoke({"messages":[HumanMessage(content="Hi What is my name")]},config=config)
```
